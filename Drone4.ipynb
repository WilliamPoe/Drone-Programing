```python
import CoDrone
from CoDrone import PlotType
import time
from CoDrone import Direction

drone = CoDrone.CoDrone()
drone.pair(drone.Nearest)

drone.set_plot_sensor(PlotType.temperature)

drone.takeoff() 

drone.set_pitch(5)
drone.move(3)

drone.set_roll(-10)
drone.move(2)

drone.set_pitch(10)
drone.move(5)

drone.turn(Direction.RIGHT,2,20)

drone.turn(Direction.RIGHT,2,30)

drone.set_pitch(5)
drone.move(2)

time.sleep(5)

drone.land()

drone.draw_plot_sensor()

drone.close()
```
